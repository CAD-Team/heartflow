```python
# Build Meshes
shape = [10,10,10]
h = 3
origin = (39.35025102748692, -182.88678578819045, 635.1077093084257)

ns = function.Namespace()
ap_topo, geom = autoplaque_img.ToNutilsMesh(shape=shape)
a = origin - h * np.ones([3])
b = origin + h * np.ones([3])
aabb = im.IM_AABB(a, b)
topo, ns.x = aabb.ToNutilsMesh(shape)
lumen_mesh.TrimByAABB(aabb)
lumen_topo, lumen_geom = lumen_mesh.ToNutilsMesh()
```

```python
bez = topo.sample('bezier', 2)
bez_proj = im.ProjectSampleTriMesh(ns.x, lumen_topo, lumen_geom, bez)
```

```python
ns.linbasis = lumen_topo.basis('spline',1)
normals = lumen_mesh.GetData("normal")
ns.Nx = ns.linbasis.dot(normals[:,0])
ns.Ny = ns.linbasis.dot(normals[:,1])
ns.Nz = ns.linbasis.dot(normals[:,2])
ns.n_i = '<Nx, Ny, Nz>_i'
ns.N = bez.asfunction(bez_proj.eval(ns.n))
```

```python
ns.Px = bez.asfunction(bez_proj.eval(lumen_geom))
ns.u_i = 'x_i - Px_i'
ns.dist = 'sqrt(u_i u_i)'
ns.d = 'u_i N_i'
ns.I = 'sign(d)'
ns.sdf = 'I dist'
```

```python
vals = bez.eval(ns.sdf)
xvals = bez.eval(ns.x)
distvals = bez.eval(ns.dist)
Ivals = bez.eval(ns.I)
nvals = bez.eval(ns.N)
```

```python
class ArrayWrapper:
    def __init__(self, array):
        self.array = array
        
class UdotV(function.Array):
    @types.apply_annotations
    def __init__(self, wrapped_array, *args:function.asarrays):
        self.array = wrapped_array.array
        shapes = set(arg.shape for arg in args)
        assert len(shapes) == 1, 'pointwise arguments have inconsistent shapes'
        shape, = shapes
        self.args = args
        super().__init__(args=args, shape=shape, dtype=np.float64)
    
    def evalf(self, x, y, z):
        return np.array([ x[i] * self.array[i][0] + y[i] * self.array[i][1] + z[i] * self.array[i][2] for i in range(len(x))])
    
    def _derivative(self, var, seen):
        return np.zeros(self.shape + var.shape)
```
